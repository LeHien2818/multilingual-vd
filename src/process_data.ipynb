```python
import os
from pathlib import Path
import pandas as pd

def load_java_dataset(base_path):
    """
    Load Java vulnerability dataset from the folder structure.
    
    Args:
        base_path: Path to the JAVA dataset folder
        
    Returns:
        pandas.DataFrame with columns: code, label, cwe_id
    """
    data = []
    base_path = Path(base_path)
    
    # Iterate through each CWE folder
    for cwe_folder in base_path.iterdir():
        if not cwe_folder.is_dir():
            continue
            
        cwe_id = cwe_folder.name  # e.g., 'CWE-074'
        
        # Iterate through 'bad' and 'good' subfolders
        for label_folder in ['bad', 'good']:
            label_path = cwe_folder / label_folder
            
            if not label_path.exists():
                continue
            
            # Determine label: 1 for bad (vulnerable), 0 for good (safe)
            label = 1 if label_folder == 'bad' else 0
            
            # Iterate through all .txt files in the folder
            for file_path in label_path.glob('*.txt'):
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        content = f.read()
                    
                    # Extract code (everything before the separator line)
                    # The separator is a line with ^^^^^ or similar pattern
                    lines = content.split('\n')
                    code_lines = []
                    
                    for line in lines:
                        if line.strip().startswith('^'):
                            break
                        code_lines.append(line)
                    
                    code = '\n'.join(code_lines).strip()
                    
                    if code:  # Only add if there's actual code
                        data.append({
                            'code': code,
                            'label': label,
                            'cwe_id': cwe_id
                        })
                        
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")
    
    # Create DataFrame
    df = pd.DataFrame(data)
    return df

# Usage
base_path = r'c:/hienlt/treesitter/dataset/JAVA'
df = load_java_dataset(base_path)

# Display info about the dataset
print(f"Total samples: {len(df)}")
print(f"\nLabel distribution:")
print(df['label'].value_counts())
print(f"\nCWE-ID distribution:")
print(df['cwe_id'].value_counts())
print(f"\nDataFrame shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
```

```python
import os
from pathlib import Path
import pandas as pd
import json

def load_dataset(base_path):
    """
    Load Java vulnerability dataset from the folder structure.
    
    Args:
        base_path: Path to the dataset folder
        
    Returns:
        pandas.DataFrame with columns: code, label, cwe_id
    """
    data = []
    base_path = Path(base_path)
    
    # Iterate through each CWE folder
    for cwe_folder in base_path.iterdir():
        if not cwe_folder.is_dir():
            continue
            
        cwe_id = cwe_folder.name  # e.g., 'CWE-074'
        
        # Iterate through 'bad' and 'good' subfolders
        for label_folder in ['bad', 'good']:
            label_path = cwe_folder / label_folder
            
            if not label_path.exists():
                continue
            
            # Determine label: 1 for bad (vulnerable), 0 for good (safe)
            label = 1 if label_folder == 'bad' else 0
            
            # Iterate through all .txt files in the folder
            for file_path in label_path.glob('*.txt'):
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        content = f.read()
                    
                    # Extract code (everything before the separator line)
                    # The separator is a line with ^^^^^ or similar pattern
                    lines = content.split('\n')
                    code_lines = []
                    
                    for line in lines:
                        if line.strip().startswith('^'):
                            break
                        code_lines.append(line)
                    
                    code = '\n'.join(code_lines).strip()
                    code = code[:-1]
                    if code:  # Only add if there's actual code
                        data.append({
                            'code': code,
                            'label': label,
                            'cwe_id': cwe_id
                        })
                        
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")
    
    # Create DataFrame
    df = pd.DataFrame(data)
    return df

```

```python
def remove_comments(code):
    code = re.sub(r'//.*?\n|/\*.*?\*/|#.*?\n', '', code, flags=re.S)
    code = re.sub(r'/\*.*?\*/', '', code, flags=re.S)
    return code
```

```python

def remove_comments_java(code):
    # Pattern explanation:
    # Group 1: Matches String literals ("...") OR Char literals ('...')
    # Match 2: Single-line comments (//...)
    # Match 3: Multi-line comments (/*...*/)
    pattern = r'("[^"\\]*(?:\\.[^"\\]*)*"|\'[^\'\\]*(?:\\.[^\'\\]*)*\')|//[^\n]*|/\*.*?\*/'
    
    # If the regex matched a string/char (Group 1), put it back. 
    # Otherwise, it matched a comment, so replace it with an empty string.
    code = re.sub(pattern, lambda m: m.group(1) if m.group(1) else '', code, flags=re.S)
    return code
```

```python
def remove_comments_python(code):
    """Remove Python comments from code."""
    code = re.sub(r'#.*?\n', '\n', code)
    # Remove docstrings (simple approach)
    code = re.sub(r'""".*?"""', '""', code, flags=re.S)
    code = re.sub(r"'''.*?'''", "''", code, flags=re.S)
    return code
```

```python
def remove_comments_generic(code, extension):
    clean_code_mapping = {
        'c': remove_comments(code=code),
        'cpp': remove_comments(code=code),
        'java': remove_comments_java(code=code),
        'py': remove_comments_python(code=code)
    }
    return clean_code_mapping.get(extension, code)
```
